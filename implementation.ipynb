```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from Pan_tompkins_algorithm import *
```

```python
# read .csv file
x = pd.read_csv('ecg_data_1.csv')

sample_rate=200
```

```python
# ecg data
ecg = x.iloc[:, 1]
```

### Call fit to execute algorithm in one step

```python
signal = Pan_tompkins(ecg, sample_rate).fit()
plt.plot(signal);
```

### Algoirhtm step wise

```python
# Raw ecg data
plt.plot(ecg);
```

```python
bandpass = Pan_tompkins(ecg, sample_rate).band_pass_filter()
plt.plot(bandpass);
```

```python
derivative_pass = Pan_tompkins(ecg, sample_rate).derivative_filter()
plt.plot(derivative_pass);
```

```python
square_pass = Pan_tompkins(ecg, sample_rate).squaring()
plt.plot(square_pass);
```

```python
integrated_signal = Pan_tompkins(ecg, sample_rate).moving_window_integration()
plt.plot(integrated_signal);
```
